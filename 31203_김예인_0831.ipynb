````python
import matplotlib
import matplotlib.font_manager as fm

fm.get_fontconfig_fonts()
font_location = 'C:/Windows/Fonts/malgunbd.ttf'
font_name = fm.FontProperties(fname=font_location).get_name()
matplotlib.rc('font', family=font_name)
````

````python
# 꺽은선형그래프
import matplotlib.pyplot as plt

mon = ['3월','4월','5월','6월','7월']
sales = [1,5,7,3,7]

plt.title('월별 판매 실적')
plt.plot(mon, sales, 'r', label='단위:개수')    # x축,y축,색상,범례
plt.legend()
plt.show()
````

## 3. 다양한 그래프
### plt.bar() : 막대그래프 - 세로형

````python
plt.title('월별 판매 실적')    # 제목
# bar 차트는 x값, y값이 모두 존재해야 한다.
plt.bar(mon, sales, color='r')
plt.show()
````

### range() 사용하기
### range(시작숫자, 종료숫자, 간격)
시작부터 종료숫자까지 간격을 두어 숫자를 선택함<br>
```
range(2,11,2)
```
2부터 11까지 간격을 2로 두고 숫자 선택<br>
-> 2, 4, 6. 8. 10

````python
# x값 range()
plt.title('월별 판매 실적')
# range(시작값, 종료값+1)
#plt.bar(range(3,8), sales, color='g')
plt.bar(range(2,11,2), sales, color='b')
plt.show()
````
