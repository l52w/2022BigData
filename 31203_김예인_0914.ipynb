```python
# 라이브러리
import matplotlib.pyplot as plt

# 한글 폰트
plt.rc ('font', family='Malgun Gothic')
```

```python
# 키와 몸무게와 체육점수의 관계
import random

height, weight, score = [],[],[]
for i in range(100) : 
    height.append(random.randint(100,200))
    weight.append(random.randint(20,100))
    score.append(random.randint(1,1000))
plt.title('키와 몸무게와 체육점수의 관계')
plt.scatter(height, weight, c=score, s=score, cmap='rainbow', alpha=0.7)
plt.colorbar(label='체육실기점수')
plt.show()
```
