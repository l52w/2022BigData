````python
# 라이브러리
import matplotlib.pyplot as plt

# 한글폰트 적용
plt.rc('font', family='Malgun Gothic')

# 꺾은선형 plt.plot
# bar 차트 plt.bar
# bar 차트 가로형 plt.barh

mon = range(3,8)
sales = [1,5,7,3,7]
plt.bar(mon, sales,color='coral', label='단위 : 개수')
plt.title('월별 판매 실적')
plt.legend()
plt.show()
````

### plt.pie() : 원그래프
```
우리 학교 100명의 혈액형을 조사한 결과가 다음과 같다.

A형 = 28명
B형 = 20명
O형 = 40명
AB형 = 12명
```

#### 원그래프 각 항목에 라벨 붙이기
```
plt.pie(리스트, labels = 라벨리스트, autopct = '%1.1f%%')
```

````python
b_type = [28, 20, 40, 12]
b_name = ['A형', 'B형', 'O형', 'AB형']

plt.title('우리 학교 혈액형 비율')
plt.pie(b_type, labels=b_name, autopct='%1.1f%%')
plt.show()
````
