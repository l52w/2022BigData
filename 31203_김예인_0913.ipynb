```python
# 라이브러리
import matplotlib.pyplot as plt
# 한글
plt.rc('font', family='Malgun Gothic' )
```

```python
month = [1,5,7,3,7]
sales = ['3월','4월','5월','6월','7월']

# 꺾은선형
plt.title('월별 판매 실적')
plt.plot(sales, month, color='orange', label='단위 : 개')
plt.legend ()
plt.show()
```

```python
# bar형 그래프
plt.bar(sales, month, color='pink')
plt.show()
```

```python
# bar형, 범례 단위 : 개수, 제목 '월별 판매 실적'
plt.title('월별 판매 실적')
plt.bar(sales, month, color='purple', label='단위 : 개수')
plt.legend()
plt.show()
```

```python
# 원형차트
b_type = [25,19,37,11]
b_name = ['A형', 'B형', 'O형', 'AB형']

plt.pie(b_type, labels=b_name, autopct='%1.1f%%')
plt.show()
```

```python
# 키와 몸무게의 상관관계
import random
height = []
weight = []
for i in range(100) :
    height.append(random.randint(100,200))
    weight.append(random.randint(20, 100))
plt.title('키와 몸무게의 상관관계')
plt.scatter(height, weight, c= height, cmap='RdPu')
plt.xlabel("키")
plt.ylabel("몸무게")
plt.show()
```

```python
# 키와 몸무게 체육실기점수의 관계
height, weight, score = [], [], []

plt.title('키와 몸무게와 실기 점수의 관계')
for i in range(100) :
    height.append(random.randint(100, 200))
    weight.append(random.randint(20,100))
    score.append(random.randint(1,1000))
plt.scatter(height, weight, s=score, cmap='rainbow', alpha=0.7)
plt.colorbar(label='체육 실기 점수')
plt.show()
```
