````python
# 라이브러리
import matplotlib.pyplot as plt

# 한글폰트
plt.rc('font', family='Malgun Gothic')
````

````python
mon = range(3,8)
sales = [1,5,7,3,7]

plt.title('월별 판매 실적')
# 꺾은선형
# plt.plot(mon, sales)
# bar chart
plt.bar (mon, sales, color='darkgreen', label='단위 : 개수')
plt.legend()
plt.show()
````

````python
# 혈액형 인원수
b_type = [28,20,40,12]
# 혈액형
b_name = ['A형', 'B형', 'O형', 'AB형']

plt.title('혈액형 비율')
plt.pie(b_type, labels = b_name, autopct='%1.1f%%')
plt.show()
````

### plt.scatter() : 산점도
> 국어를 잘하는 친구가 영어도 잘할까?
```
국어점수와 영어점수의 상관관계 알아보기
kor = [ 80, 20, 50, 20, 10, 50, 60, 30, 60 ]
eng = [ 90, 40, 60, 40, 10, 30, 50, 70, 90 ]
```

````python
kor = [ 80, 20, 50, 20, 10, 50, 60, 30, 60 ]
eng = [ 90, 40, 60, 40, 10, 30, 50, 70, 90 ]

plt.scatter(kor, eng)
plt.show()
````

````python
plt.title('국어 점수와 영어 점수의 상관관계')
plt.scatter(kor, eng, color='indigo')

# x, y축 라벨
plt.xlabel('국어 점수')
plt.ylabel('영어 점수')

plt.show()
````
